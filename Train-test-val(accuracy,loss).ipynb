```python
import pickle
import matplotlib.pyplot as plt

# Load training history from pickle
history_file = 'inceptionResnetv2_GRU_new.pkl'
with open(history_file, 'rb') as f:
    history = pickle.load(f)

# Extract metrics
acc = history['accuracy']
val_acc = history['val_accuracy']
loss = history['loss']
val_loss = history['val_loss']
epochs = range(1, len(acc) + 1)

# External validation results
external_acc = 0.8373
external_loss = 0.5234

# Best epoch 
best_epoch = 12

# Plot Accuracy
plt.figure(figsize=(6, 5))
plt.plot(epochs, acc, color='orange', label='Training Accuracy', marker='o')
plt.plot(epochs, val_acc, 'g-', label='Validation Accuracy', marker='o')
plt.axhline(y=external_acc, color='r', linestyle='--', label='External Validation Accuracy')
plt.axvline(x=best_epoch, color='purple', linestyle='--', label='Best Epoch')
plt.title('InceptionResNetV2 + GRU Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Plot Loss
plt.figure(figsize=(6, 5))
plt.plot(epochs, loss, color='orange', label='Training Loss', marker='s')
plt.plot(epochs, val_loss, 'g-', label='Validation Loss', marker='s')
plt.axhline(y=external_loss, color='r', linestyle='--', label='External Validation Loss')
plt.axvline(x=best_epoch, color='purple', linestyle='--', label='Best Epoch')
plt.title('InceptionResNetV2 + GRU Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import pickle
import matplotlib.pyplot as plt

# Load histories
with open('inceptionResnetv2_GRU_new.pkl', 'rb') as f:
    history_gru = pickle.load(f)

with open('RN50hp_new_5.pkl', 'rb') as f:
    history_resnet = pickle.load(f)

# Extract metrics
acc_gru = history_gru['accuracy']
val_acc_gru = history_gru['val_accuracy']
loss_gru = history_gru['loss']
val_loss_gru = history_gru['val_loss']

acc_resnet = history_resnet['accuracy']
val_acc_resnet = history_resnet['val_accuracy']
loss_resnet = history_resnet['loss']
val_loss_resnet = history_resnet['val_loss']

epochs = range(1, len(acc_gru) + 1)

# External values
external_acc_gru = 0.8373
external_loss_gru = 0.5234
external_acc_resnet = 0.82
external_loss_resnet = 0.5921

# Best epochs
best_epoch_gru = 12
best_epoch_resnet = 18

# --- Accuracy Plot ---
plt.figure(figsize=(7, 6))

# GRU
plt.plot(epochs, acc_gru, color='orange', linewidth=2, label='GRU - Train Acc', marker='o')
plt.plot(epochs, val_acc_gru, color='orange', linestyle='--', linewidth=2, label='GRU - Val Acc', marker='o')
plt.axhline(y=external_acc_gru, color='red', linestyle=':', linewidth=2, label='GRU - External Acc')
plt.axvline(x=best_epoch_gru, color='purple', linestyle=':', linewidth=2, label='GRU- Best epoch')

# ResNet50
plt.plot(epochs, acc_resnet, color='darkgreen', linewidth=2, label='ResNet50 - Train Acc', marker='s')
plt.plot(epochs, val_acc_resnet, color='darkgreen', linestyle='--', linewidth=2, label='ResNet50 - Val Acc', marker='s')
plt.axhline(y=external_acc_resnet, color='navy', linestyle=':', linewidth=2, label='ResNet50 - External Acc')
plt.axvline(x=best_epoch_resnet, color='darkgreen', linestyle=':', linewidth=2, label='ResNet- Best epoch')

plt.title('Accuracy Comparison (GRU vs ResNet50)', fontsize=14)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='lower right', fontsize=9)
plt.tight_layout()

# --- Loss Plot ---
plt.figure(figsize=(7, 6))

# GRU
plt.plot(epochs, loss_gru, color='orange', linewidth=2, label='GRU - Train Loss',marker='o')
plt.plot(epochs, val_loss_gru, color='orange', linestyle='--', linewidth=2, label='GRU - Val Loss',marker='o')
plt.axhline(y=external_loss_gru, color='red', linestyle=':', linewidth=2, label='GRU - External Loss')
plt.axvline(x=best_epoch_gru, color='purple', linestyle=':', linewidth=2, label='GRU- Best epoch')

# ResNet50
plt.plot(epochs, loss_resnet, color='darkgreen', linewidth=2, label='ResNet50 - Train Loss', marker='s')
plt.plot(epochs, val_loss_resnet, color='darkgreen', linestyle='--', linewidth=2, label='ResNet50 - Val Loss', marker='s')
plt.axhline(y=external_loss_resnet, color='navy', linestyle=':', linewidth=2, label='ResNet50 - External Loss')
plt.axvline(x=best_epoch_resnet, color='darkgreen', linestyle=':', linewidth=2, label='ResNet- Best epoch')

plt.title('Loss Comparison (GRU vs ResNet50)', fontsize=14)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right', fontsize=9)
plt.tight_layout()

plt.show()
```
